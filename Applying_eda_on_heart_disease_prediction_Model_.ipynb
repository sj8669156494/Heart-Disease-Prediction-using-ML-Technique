```python
import numpy as np 
import pandas as pd 
import os
```

```python
heart_df = pd.read_csv("/content/heart_disease_data-2.csv")
heart_df.head()
```

```python
heart_df.target.value_counts()
```

```python
# Observe the counts in percentages 
heart_df.target.value_counts(normalize = True) # normalize parameter
```

```python
import matplotlib.pyplot as plt
heart_df.target.value_counts().plot(kind = "bar", color = ["red", "blue"])
```

```python
heart_df.sex.value_counts()
```

```python
pd.crosstab(heart_df.target, 
            heart_df.sex)
```

```python
pd.crosstab(heart_df.target, 
            heart_df.sex, normalize = True)
```

```python
pd.crosstab(heart_df.target, 
            heart_df.sex).plot(kind ="bar", figsize = (10,5), color = ["red","blue"])

plt.title("heart disease frequency based on sex")
plt.xlabel("0 = female 1 = male")
plt.ylabel("Count")
plt.legend(["Female", "Male"])
plt.xticks(rotation = 0);
```

```python
plt.figure(figsize = (10,6))

plt.scatter(heart_df.age[heart_df.target == 1],
           heart_df.chol[heart_df.target ==1], 
           color = "red")

plt.scatter(heart_df.age[heart_df.target==0],
           heart_df.chol[heart_df.target==0],
           color = "blue")

plt.xlabel("age")
plt.ylabel("cholestrol")
plt.legend(["Disease", "No Disease"])
plt.title("Heart disease on age and cholestrol level")
plt.xticks(rotation = 0);

```

```python
pd.crosstab(heart_df.target, heart_df.cp) # cp = Chest Pain
```

```python
pd.crosstab(heart_df.cp,heart_df.target).plot(kind = "bar", figsize = (10, 6), color = ["red", "blue"])

plt.title("Heart disease based on chest pain")
plt.xlabel("Chest Pain")
plt.ylabel("Heart Disease")
plt.legend(["Disease", "No Disease"])
```
